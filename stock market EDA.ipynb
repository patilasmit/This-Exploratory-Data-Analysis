```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import pandas_datareader as pdr
```

```python
stock = 'RELIANCE.NS'
rel_data = yf.download(stock,start = '2022-01-01' ,end = '2024-01-01')
```

# plot the closing price 

```python
plt.figure(figsize = (15,5))
sns.lineplot(data=rel_data ,x='Date' , y='Adj Close')
plt.title(f'{stock} Closing Price')
plt.show()
```

# Histogram of Returns 

```python
rel_data['Returns'] = rel_data['Adj Close'].pct_change()
```

```python
plt.figure(figsize = (25,8))
sns.lineplot(data=rel_data ,x='Date' , y='Returns')
plt.title(f'{stock} Returns')
plt.xlabel('Returns')
plt.show()
```
